```python
#Predict categorical target variable (malign or benign cancer) (source: Simplilearn youtube course) 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns; sns.set(font_scale=1.2)
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
data     = pd.read_csv('diagnostics.csv')
data     = data[data.columns[1:]]
y        = data[data.columns[0]]  #target
data     = data[data.columns[1:]]
data     = data.apply(pd.to_numeric, errors='coerce') # convert all columns of DataFrame to numeric data
#handling missing data
for i in data.columns:
    data[i] = data[i].fillna(int(data[i].mean()))
```

```python
#check if there is null data
data.isnull().sum() #if != 0, it means that there is at least 1 Nan value
```

```python
# jointplot
c1 = 0
c2 = 2
sns.jointplot(x=data.columns[c1], y=data.columns[c2], data=data)
```

```python
# correlation heatmap
sns.heatmap(data.corr())
```

```python
# dividing the data (getting only the 'worst')
worst_columns = [c for c in data.columns.to_list() if "worst" in c]
x = data[worst_columns]
```

```python
#data train/test spliting
from sklearn.model_selection import train_test_split
t_size = 0.7
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=t_size,random_state=101)
```

```python
#training logistic regression (good for categorical)
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
fit   = model.fit(x_train,y_train)
```

```python
#obtain the peformance of the model using classification_report metrics
from sklearn.metrics import classification_report
y_pred = model.predict(x_test)
report = classification_report(y_test,y_pred)
print(report)
```
